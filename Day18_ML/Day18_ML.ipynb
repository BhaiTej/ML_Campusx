## Seaborn

```python
f=pd.read_csv('train.csv')
df=pd.DataFrame(f)
df
```

# 1.Categorical Data

## a. Countplot

```python
# sns.countplot(df['Survived'])

# You're passing a Series directly as the first positional argument. 
# While older versions of Seaborn may have tolerated this
# , newer versions expect you to explicitly specify the axis (x or y) 
# and use the data parameter.

sns.countplot(x=df['Survived'],data=df)
# df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(x=df['Pclass'],data=df)
```

```python
sns.countplot(x=df['Sex'],data=df)
```

```python
# Embarked
sns.countplot(x=df['Embarked'],data=df)
```

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```
